```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

Loading embeddings

```python
em = np.load("/mnt/ricproject4/anuraj_works/git_repo/voxtrainer_IV/embeddings/embeddings_english.npy")
em2 = np.load("/mnt/ricproject4/anuraj_works/git_repo/voxtrainer_IV/embeddings/embeddings_sinhala.npy")
em3 = np.load("/mnt/ricproject4/anuraj_works/git_repo/voxtrainer_IV/embeddings/embeddings_tamil.npy")
```

Ploating the PCA

```python
x = em[:,1:513] # loading the featuers
y = em[:,0] # loading the speaker label
x2 = em2[:,1:513]
y2 = em2[:,0]
x3 = em3[:,1:513]
y3 = em3[:,0]

x = StandardScaler().fit_transform(x)
x2 = StandardScaler().fit_transform(x2)
x3 = StandardScaler().fit_transform(x3)

pca = PCA(n_components = 2)
pcs = pca.fit_transform(x)
pcs2 = pca.fit_transform(x2)
pcs3 = pca.fit_transform(x3)

pcadfE = pd.DataFrame(data = pcs, columns = ['pcaE1', 'pcaE2'])
pcadfT = pd.DataFrame(data = pcs2, columns = ['pcaT1', 'pcaT2'])
pcadfS = pd.DataFrame(data = pcs3, columns = ['pcaS1', 'pcaS2'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(40,40))

plt.scatter(
    x=pcadfE.pcaE1,
    y=pcadfE.pcaE2,
    color="red",
    label ="English",
)

plt.scatter(
    x=pcadfT.pcaT1,
    y=pcadfT.pcaT2,
    color="green",
    label = "Tamil",
)
plt.scatter(
    x=pcadfS.pcaS1,
    y=pcadfS.pcaS2,
    color="blue",
    label = "Sinhala",
)
```

Ploating the TSNE

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)

x = em[:,1:513]
x2 = em2[:,1:513]
x3 = em3[:,1:513]

tsne_results1 = tsne.fit_transform(x)
tsne_results2 = tsne.fit_transform(x2)
tsne_results3 = tsne.fit_transform(x3)

#Defining the TSNE plot axis
tsneE1 = tsne_results1[:,0]
tsneE2 = tsne_results1[:,1]

tsneS1 = tsne_results2[:,0]
tsneS2 = tsne_results2[:,1]

tsneT1 = tsne_results3[:,0]
tsneT2 = tsne_results3[:,1]
```

Drawing TSNE plot

```python
plt.figure(figsize=(20,20))
sns.scatterplot(
    x=tsneE1, y=tsneE2,
    palette=sns.color_palette("hls", 10),
    legend="full",
    alpha=0.3,
    color="green",
)

sns.scatterplot(
    x=tsneS1, y=tsneS2,
    palette=sns.color_palette("hls", 10),
    legend="full",
    alpha=0.3,
    color="red",
)

sns.scatterplot(
    x=tsneT1, y=tsneT2,
    palette=sns.color_palette("hls", 10),
    legend="full",
    alpha=0.3,
    color="blue",
)
```
